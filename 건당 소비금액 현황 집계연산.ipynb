```python
import pandas as pd
```

```python
df = pd.read_csv("(평균)유성구 건당 평균 소비 데이터.csv")
df.head()
```

```python
df1 = df.iloc[:,2:6]
df2 = df.iloc[:,16:]
```

```python
df3 = pd.concat([df1, df2], axis=1)
df3
```

```python
# 확인 1
df3.query("법정동명 == '방동'")
```

```python
# 확인 2
df4 = df3.groupby('표준산업대분류명',as_index=False).mean()
df4.T
```

```python
df4 = df3.query("법정동명=='방현동'").groupby(['법정동코드', '법정동명'],as_index=False).mean()
df4
```
