### lowss 극소회귀
- 가중치 함수는 추정 지점에 가장 가까운 데이터 포인트에 가장 많은 가중치를 부여하고 가장 먼 데이터 포인트에 가장 적은 가중치를 부여한다.
- lowess: locally weighted robust scatterplot smoothing

```python
#loswss 회귀
sns.lmplot(x='1인당 평균 소비 금액', y = 'MZ세대 평균 소비 금액', data = df,
          lowess=True)
```

```python
#gridspec 이 모듈을 가지고 온다!
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) #본인이 그리고 싶은 그래프의 개수를 적으면 됩니다.
plt.figure(figsize=(15,10))
plt.subplots_adjust(wspace=0.5, hspace=0.3) #현재 배치되어 있는 subplot들의 위치를 직접 조정한다
                                            # wspace 가로 간격, hspace 세로 간격

    
#서브플롯에 필요한 컬럼을 가져온다!
mpg_features = ['M세대인구비율','Z세대인구비율','아파트수 비율','1인당 평균 소비 금액']

for idx, feature in enumerate(mpg_features): # 인덱스(index)와 원소를 동시에 접근하면서 루프를 돌릴 수 있다.
    ax = plt.subplot(grid[idx])
    
    sns.countplot(x=feature,      # feature calss
                 data=df,        # data는 mpg 데이터
                 hue='MZ세대 평균 소비 금액',       # drv 컬럼의 값별로 count
                 palette='pastel',# 색 구성
                 ax=ax)           # 그래프위 위치 결정
    ax.set_title(f'{feature} Distribution')
```
